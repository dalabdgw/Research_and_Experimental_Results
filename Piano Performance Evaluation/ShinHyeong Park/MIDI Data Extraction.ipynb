#### (2). MIDO 데이터 설명하기

- MidiFile(type=1, ticks_per_beat=480, tracks=[...]):
    - type=1 : 여러 track 을 가질 수 있으며 각 트랙은 자체 음악 이벤트 세트를 가짐
    - ticks_per_beat=480 : MIDI의 시간 단위인 tick 당 분해를 의미, 타이밍의 해상도
    - tracks : MIDI 트랙
        * MetaMessage('track_name') : 트랙의 이름
        * MetaMessage('set_tempo', tempo=500000, .. ) : 트랙의 템포, 500,000 마이크로초당 4분음표
        * MetaMessage('time_signature', numerator=4, denominator=4, ... ) : 트랙의 박자, 4/4 박자
        * 이벤트 ( 3가지 )
            - ▫️ Message('note_on', channel=0, note=45, velocity=78, time=64), 
                * (1). 'note on' : 음표를 연주, 소리를 On
                * (2). channel = 0 : MIDI 채널을 나타냄, 0번 채널
                * (3). note = 45 : 연주할 음표의 MIDI 노트 넘버
                * (4). velocity = 78 : 연주되는 음표의 강도 또는 속도 // 강도에 대한 정보
                * (5). time = 64 : MIDI 이벤트가 트랙에 삽입된 시간 ( 단위 : tick ), 64 tick 이후 발생
                * (6). control = MIDI 컨트롤러 조작 (ex. control=0 // 전체 볼륨 조절 )
                * (7). value = 컨트롤러에 할당된 값
            - ▫️ Message('note_off', channel=0, note=45, velocity=64, time=152), // 연주 음표가 멈춤
            - ▫️ Message('aftertouch', channel=0, value=4, time=291), // MIDI 키보드에서 손가락 압력을 감지

- MIDI 데이터
    - 트랙 이름
    - 트랙 정보 : ( 템포, 박자, 총 시간 ( tick ) )
    - 소리 정보 : ( MIDI 노트 넘버, 채널, 강도, 시간 )
    

- 부가 설명
    - MIDI 음악 이벤트
        - note_on : 특정 음표를 연주하거나, 소리를 켜는 것, 보통 새로운 음표의 시작
        - aftertouch : 이미 연주 중인 음표에 대한 압력 변화, 
    - track
    - channel
        - 16개의 채널을 가짐
        - 하나의 트랙 내에서 (( 여러 채널 )) 을 사용하여, 동시에 여러 악기를 연주
        - 
    - 악기 연주가 바뀌는 순간
        - Message('program_change', channel=2, program=26, time=0),// program_change 로 알 수 있음
        - program : 0 ~ 127
        - General MIDI Level 1의 0부터 127까지의 MIDI 프로그램 넘버에 대응하는 일반적인 악기
            - 0   Acoustic Grand Piano
            - 4   Electric Piano 1
            - 5   Electric Piano 2
