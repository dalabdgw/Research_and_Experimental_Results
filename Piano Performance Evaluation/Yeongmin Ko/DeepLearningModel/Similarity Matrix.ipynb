|--|--|--|--|--|--|--|
|딥러닝 모델 1|45|64|66|22.4|45.1|10|
|딥러닝 모델 2|13|46|44|22|17|25|
|알고리즘|9|8|7|6|1|0|

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = np.array([
    [45, 64, 66, 22.4, 45.1, 10],
    [13, 46, 44, 22, 17, 25],
    [9, 8, 7, 6, 1, 0],
])

similarity_matrix = np.dot(data, data.T) / (np.linalg.norm(data, axis=1)[:, None] * np.linalg.norm(data, axis=1))

print(similarity_matrix)
```

```python
#plt.imshow(similarity_matrix, cmap='hot', vmin=0, vmax=1)
plt.imshow(similarity_matrix, cmap='viridis', vmin=0, vmax=1)
plt.colorbar()

names = ['Deep 1', 'Deep 2', 'Algo 1']

plt.xticks(np.arange(len(names)), names)
plt.xlabel('Model')

plt.yticks(np.arange(len(names)), names)
plt.ylabel('Model')

plt.show()
```
