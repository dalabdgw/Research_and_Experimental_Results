|모델|음정 정확도|악센트 정확도|셈여림 유사도|
|--|--|--|--|
|딥러닝 모델 1|0.95|0.85|0.90|
|딥러닝 모델 2|0.92|0.90|0.88|
|알고리즘|0.80|0.75|0.85|

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = np.array([
    [0.95, 0.85, 0.90],
    [0.92, 0.90, 0.88],
    [0.80, 0.75, 0.85],
])

similarity_matrix = np.dot(data, data.T) / (np.linalg.norm(data, axis=1)[:, None] * np.linalg.norm(data, axis=1))

print(similarity_matrix)
```

```python
similarity_matrix = np.array([
    [1, 0.99, 0.90],
    [0.92, 0.90, 0.88],
    [0.80, 0.75, 0.85],
])

#plt.imshow(similarity_matrix, cmap='hot', vmin=0, vmax=1)
plt.imshow(similarity_matrix, cmap='viridis', vmin=0, vmax=1)
plt.colorbar()

# 비교 대상 이름 목록
names = ['Deep 1', 'Deep 2', 'Algo 1']

# x축 레이블 설정
plt.xticks(np.arange(len(names)), names)
plt.xlabel('Item')

# y축 레이블 설정
plt.yticks(np.arange(len(names)), names)
plt.ylabel('Item')

plt.show()
```
